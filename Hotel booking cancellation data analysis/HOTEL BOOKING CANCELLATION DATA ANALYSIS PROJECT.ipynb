**- 3rd ANALYSIS :** Kya price ka koi affect hai dono hotel ki cancelation me ?

```python
# GROUP BT FUNCTION ka use krenge Group krne k liye on the 'reservation_status_date'
# Kyuki 1 hi din k bohot sare records hai
# To 1 din ke AVERAGE DAILY RATE ka mean value lenge

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# - Now we will create a visualization for the AVERAGE DAILY RATE in both city and resort hotels
# Hmne hypothesis me esa dekha tha ki jo RESORT HOTEL hai usme cancelation ke chances PRICE k vjh se hai
# To use chiz ko check krenge

plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City & Resort Hotel (ADR)' , fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()

# X value will be the Index (Index is reservation_status_date)
# Y will be the AVERAGE DAILY RATE




# ####################################################
# City hotel ki lines mid me hai ==> What does that mean ?? ==> 
# Or kuch spikes bhi hai ==> WHY? ==> Spikes is for weekends...weekends pe dono hotels ka price increase ho jata hai.
# City hotel ka price as compared to resort hotel km hai.
# Hypothesis is proved ==> City hotel price < resort hotel price
```

### Findings :
1. City hotel is in mid (ADR) , some spikes may be due to weekends,holidays
2. City hotel price < resort hotel price

HENCE, HYPOTHESIS IS PROVED !!

**- 4th ANALYSIS :** Kis month me sbse jyada reservations hai or kisme sb se jyada cancelations .

```python
# We'll use bar chart to see the cancelations and not cancelations in a month by grouping then we can compare 
# We want x axis pe month ==> for that ==> hmare pass month ka koi column nhi hai reeservation_status_date ke liye 
# Hmne use datetime m convert kiya tha ab usme month krke 1 date extend krna hai

df['month'] = df['reservation_status_date'].dt.month  # ==> Sbse pehle month kkrke 1 column add kiya....jitni bhi date hai uske month isme return ho jayenge.
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month' , hue = 'is_canceled' , data = df)
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation Status Per Month' , size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled' , 'canceled'])
plt.show()


# x-axis me sare month k number h 1-12
# January me sbse jyada cancelations hue hai
# Sbse lowest hue hai Auguest me
# Analysis says that :
# Jb cancelations sbse km hai ===> tb hmare reservations sbse jyada hai
# Or jb cancelations sbse jyada hai ===> tb hmare reservation k hai
# Mtlb === > ?????????????
# CASE - 1 : Auguest me hotels k price km ho skte hai is vajah se jagah nhi mili ==> long waiting list ==> RESERVATION CANCELED
# OR
# CASE - 2 : Ya to price bohot jyada the to logo ne bookings cancel krva di
```

### Findings :
1. Jan has more cancelations
2. Lowest cancelations in auguest
3. Auguest has most reservations
4. December Jan has least reservations.

### Bit Confusing !
Aug has least cancelations and most reservations , may be the prices are low in Aug so more reservations , or maybe the prices were too high , that people preferred cancelations.

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'not canceled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'canceled')
plt.legend
plt.show()


# data jyada hai isliye esa graph hai
```
