```python
#Now we will create a chart on the impact of parents' education on student scores.
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore" :'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean'})                  
#gb >> group by ...

#.agg({}) >> We will use this to calculate the mean value of students' math scores, reading scores, and writing scores. 
#We will pass these three things in the format of a dictionary.

#Now we will print all the values.

print(gb)
```

```python
#All our values have been printed.
```

```python
#Now looking at the data, it's clear that students whose parents have bachelor's or master's degrees have higher scores.

#Now we'll plot this effectively.
```

# Heatmap

```python
sns.heatmap(gb)   #We have a heat map within Seaborn >>> where we will pass the GB.

plt.title("Relationship between Parent Education and students score")  
#This will be the title of the chart, which will tell us what the chart is about.

plt.show()
```

```python
#Now from the above map we are understanding that for lower values there is a darker shade and for lighter shades there are higher values.
#A color scale is given on the side so we are understanding this thing by looking at it.
```

```python
#Now only colors are showing, we also want to show values, so we will...

sns.heatmap(gb,annot=True)   #In Seaborn, we have a heatmap >>> in which we will pass the gb.

plt.title("Relationship between Parent Education and students score")  
#This will add a title to the chart, which will tell us what the chart is about.

plt.show()



#As soon as we run this, the values started showing to us.
```

```python
#Now, if we want, we can also change the size of this in the same method as we did in the gender distribution chart.
```

```python
plt.figure(figsize=(5,5))

plt.title("Relationship between Parent Education and students score")  
#This will add a title to the top of the chart, indicating what the chart is about.


sns.heatmap(gb,annot=True)   #We have a heatmap within Seaborn >>> we will pass the data into it.
plt.show()
```

```python
#From the above chart, we have concluded that parents' education has a positive impact on students' scores.
```

# parents maritial status on students score 

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore" :'mean',"ReadingScore" : 'mean',"WritingScore" : 'mean'})
print(gb1)
```

```python
# If you want to create a chart for this, we can use seaborn to create a heatmap chart.

plt.figure(figsize=(5,5))

plt.title("Relationship between Parent Marital Status and students score")  
#This will be the title of the chart, which will tell us what the chart is about.

sns.heatmap(gb1,annot=True)   #We have a heat map in seaborn >>> where we will pass the GB. 
plt.show()
```

```python
#The parents' marital status is not having a significant impact on students' results.
```

```python
#So from the above chart, we have analyzed that there is no/negligible impact on students' scores due to their parents' marital status.
```

# Distribution of Ethnic Group

```python
# We will use a pie chart for finding out percentages.

# We will use the .loc function of pandas.
```

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
print(groupA)

#We only have the values or details that are present within Group "A".
```

```python
#We can check this very thing by removing the parentheses after the count.
groupA=df.loc[(df["EthnicGroup"]=="group A")].count
print(groupA)

```

```python
#Now we can also calculate for the remaining groups.
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()






l = ["groupA","groupB","groupC","groupD","groupE"]

mlist=(groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"])
plt.pie(mlist,labels=l , autopct="1.2f%%")               #DataHasBeenPlotted #GraphCreated.     #LabelsAdded #GroupNamesAssigned. 
                                                        #ShowPercentages #PercentageValuesInChart. 


    
plt.title("Distribution of Ethnic Group")   # We used it to put in the title..  
plt.show()


print(mlist)    #In the end, we also printed and checked the values.
```
